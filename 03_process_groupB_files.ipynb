# Pipeline

```
for FASTQ in $(ls $RAW_FASTQS | grep ^B | head -2); do
    FASTQ_BASE=$(basename ${FASTQ} '_001.fastq.gz')
    
    echo $FASTQ
    echo $FASTQ_BASE
    echo ""
    ##################################################
    fastqc --extract $RAW_FASTQS/B_16_S26_L001_R1_001.fastq.gz -o $QC
    
    fastq-mcf $ADAPTERS \
        $RAW_FASTQS/B_16_S26_L001_R1_001.fastq.gz \
        -q 20 -x 0.5 \
        -o $TRIMMED/B_16_S26_L001_R1_001.trim.fastq.gz
        
    fastqc --extract $RAW_FASTQS/B_16_S26_L001_R1_001.fastq.gz -o $QC
    
    STAR \
        --runMode           alignReads  \
        --twopassMode       None        \
        --genomeDir         $GENOME_DIR \
        --readFilesIn       $TRIMMED/B_16_S26_L001_R1_001.trim.fastq.gz \
        --readFilesCommand  gunzip -c                                   \
        --outFileNamePrefix ${STAR_OUT}/B_16_S26_L001_R1_short_introns_ \
        --quantMode         GeneCounts              \
        --outSAMtype        BAM SortedByCoordinate  \
        --alignIntronMax    5000                    \
        --outSJfilterIntronMaxVsReadN 500 1000 2000
        
   echo "---------------- INDEXING BAM: $FASTQ_BASE ----------------"
   samtools index ${STAR_OUT}/${FASTQ_BASE}_short_introns_Aligned.sortedByCoord.out.bam
done
```

````bash
for file in ${STAR_OUT}/*_short_introns_Log.final.out; do
    echo $(basename ${file} "_short_introns_Log.final.out") 
    cat ${file} | grep "Number of input reads" | awk '{$1=$1;print}'
    echo ""
done
````
